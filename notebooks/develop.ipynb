### Introdccion al PCA
Principal Component Analysis (PCA) es un método de reducción de dimensionalidad que se utiliza para reducir el número de variables en un conjunto de datos.

En machine learning es normal encontrarnos con problemas donde tengamos una enorme cantidad de variables con relaciones complejas entre ellas y con la variable objetivo. En estos casos, PCA puede ser una herramienta muy útil para reducir la cantidad de variables a un número manejable, manteniendo la mayor cantidad de información posible.

**Cuando usarlo**:
* Cuando tengamos un dataset con muchas variables 
* Cuando queramos reducir el costo computacional de nuestro modelo
* Hay una alta correlación entre las variables
* Se está teniendo un overfitting

Pasos del analisis de componentes principales
1. Calcular la nmatrix de covarianza para expresar las relaciones mas fueres entre las variables.
2. Calcular los autovectores y autovalores de la matriz de covarianza. Para medir la fuerza y la variabilidad de estas relaciones.
3. Ordenar los autovalores de mayor a menor y seleccionar los k autovectores que corresponden a los k autovalores más grandes. es decir, que brindan la mayor cantidad de información.  

#### Variantes del PCA:
* **Kernel PCA**: Es una versión no lineal de PCA. Se utiliza cuando los datos no son linealmente separables. por lo que se procesan con un Kernel.
* **Incremental PCA**: Se utiliza cuando el dataset es muy grande y no cabe en memoria.

## Regularizacion
La regularización es una técnica utilizada para evitar el sobreajuste en los modelos de Machine Learning. 

* Consiste en disminuir la complejidad del modelo a través de una penalización aplicada a sus variables mas irrelevantes.
* Consiste en cambiar:  + sesgo por - varianza

Para poder aplicar la regularización a un modelo, se debe definir el concepto de perdida, Una medida que nos dice qué tan lejos se está de los valores reales.  **En el conjunto de validacion** se debe tomar el minimo, porque cuando empieza a subir es porque esta sobreajustando.



### Algunas técnicas de regularización
* **L1 (Lasso)**: Se utiliza para seleccionar variables importantes y eliminar las irrelevantes.
* **L2 (Ridge)**: Se utiliza para reducir la complejidad del modelo. sin eliminar los features irrelevantes ya que pueden ayudar a regularizar a largo plazo.
* **Elastic Net**: Es una combinación de L1 y L2. Se utiliza cuando se tienen muchas variables y se quiere seleccionar las más importantes.

## Datos Atípicos
Los datos atípicos son valores que se alejan significativamente de los demás valores en un conjunto de datos.

Pueden indicar:
* Errores en la recolección de datos
* Variabilidad en un sistema
* Cambios en la distribución de los datos (novedades)

**¿Por qué son un problema?**
1. Pueden afectar la precisión de los modelos de Machine Learning.
2. Pueden contener información valiosa acerca de la naturaleza de los datos.
3. Pueden ayudar a la deteccion temprana de fallos del modelo.

**Cómo detectarlos**
* Z-Score: mide la distancia en desviaciones estándar de un punto de datos con respecto a la media.
* Tecnicas de clustering como el DBSCAN
* Rango intercuartilico
    * Visualización de los datos con boxplots

    


## Regresiones Robustas en Scikit-Learn
Algunos modelos permiten tratar directamente los datos atipicos por medio de los metaestimadores, y estas técnicas son conocidas como regresiones robustas

* **RANSAC**: Random Sample Consensus. 
    * Selecciona un subconjunto aleatorio de los datos y ajusta un modelo a estos datos.
    * Los puntos que están cerca del modelo se consideran inliers y los que están lejos se consideran outliers.
    * Se repite el proceso varias veces y se selecciona el modelo que tiene más inliers.
    * Es muy útil cuando se tienen muchos datos atípicos.

    * RECURSO : https://www.youtube.com/watch?v=9D5rrtCC_E0&ab_channel=CyrillStachniss

* **Huber Regressor**: No intenta sacar los valores atipicos sino penalizarlos.
    * Los datos son tratados como atipicos si el error absoluto de nuestra perdida está por encimade un umbral epsilon.
    * Se ha demostrado que un valor epsilon de 1.35 es optimo para la mayoria de los casos.

# Métodos de ensamble
* Combina varios modelos, logrando un consenso entre ellos, para mejorar la generalización y el rendimiento del modelo.
* La diversidad de los modelos es clave para el éxito de los métodos de ensamble.
* Se han destacado por su eficacia en competencias de Machine Learning. **!!**

**Algunas Estrategias :**
* **Bagging**: ¿y si en lugar de tener la opinion de un experto logramos tener el consenso de varios?  **Trabajo en paralelo**
    * Bootstrap Aggregating. Consiste en entrenar varios modelos con diferentes subconjuntos de datos y promediar sus predicciones.
    * Modelos ensamblados en Bagging:
        * Random Forest
        * Voting classifier/Regressors
        * Se puede aplicar sobre cualquier familia de machine learning

* **Boosting**: Le pedimos a un experto su opinion, y le pedimos al siguiente experto que nos brinde su opinon basandose en el error el anterior, y así ... **Trabajo en secuencia**
    * Viene de impulsar / propulsar
    * Busca fortalecer modelos de menor complejidad a partir de la repericion secuencial
    * Finalmente el resultado tambien se está consiguiendo por consenso entre todos los modelos.
    * Modelos ensamblados en Boosting:
        * AdaBoost
        * Gradient Boosting

# Validacion de Modelos (Cross Validation)

**¿Qué es la validación?**
* La ultima palabra siempre la tienen los datos
* Mentalidad de testeo continuo, con distintas configuraciones, de datos, modelos, parametros, etc.
* Todos los modelos son malos, solo que algunos son útiles.

**Tipos de validacion**
1. **Holdout**: Se divide el dataset en tres partes, Train, Validation y Test. para cada combinacion de parametros se entrena con Train y se valida con Validation. luego, al escoger la metrica de mejor rendimiento se testea con Test.

2. **K-Fold**: Se divide el dataset en k partes iguales, se entrena con k-1 y se testea con 1.
    * Se divide en k partes
    * Se entrena con k-1 y se testea con 1
    * Se repite k veces
    * Luego se promedian los resultados
    * se hace para cada combinacion de parametros
    * Se escoge la mejor combinacion de parametros
    * Ahora se entrena toooodo el dataset con esa combinacion de parametros y se testea con el dataset de test.

3. **Loocv**: Leave One Out Cross Validation. Se entrena con n-1 y se testea con 1. **PERO** Se repite n veces. (es como el cross validation pero con k=n, y n es el numero de datos)

**Cuando utilizar cada uno**
* Holdout: 
    * se requiere prototipado rapido
    * no se tiene mucho conocimiento del dataset
    * No se cuenta con abundantes recursos computacionales

* K-Fold CV :
    * Recomendado para la mayoria de los casos
    * Se cuenta con equipo suficiente
    * Se requiere la integracion con tecnicas de optimizacion parametrica
    * Se tiene mas tiempo para las pruebas

* LOOCV:
    * Gran poder de computo
    * Se tiene un dataset pequeño como para dividirlo en Train y Test
    * Personas que sufren de toc y quieren probar todos lo datos posibles


# Optimizacion de Hiperparametros
* **Hiperparametros**: Son parametros que no se aprenden directamente dentro de los estimadores. Son parametros que se ajustan antes de ajustar el modelo.

* Es facil perderse entre los conceptos de tantos parametros
* Es dificil medir la sensibilidad de los mismos manualmente
* Es muy costoso computacionalmente

**3 Enfoques**
* **Busqueda Manual**:
    1. Escoger el modelo que se quiere ajustar
    2. Buscar los hiperparametros que se quieren ajustar
    3. Identificar los posibles ajustes
    4. Probar las combinaciones una a una evaluando a través de listas su rendimiento.

* **Grid Search**:
    * Se define una grilla de hiperparametros, que tiene todas las combinaciones posibles de los hiperparametros que se quieren ajustar y los posibles valores que pueden tomar.
    * Usar cross validation para evaluar el rendimiento de cada combinacion de hiperparametros.
    * Entrenar el modelo e ir por un café, porque esto puede tardar un rato.

* **Random Search**:
    * Definir la métrica que se quiere optimizar
    * Identificar los rangos que los parámetros pueden tomar
    * Crear una grilla de rangos para los parámetros
    * usar cross validation para evaluar el rendimiento de cada combinacion de hiperparametros.
